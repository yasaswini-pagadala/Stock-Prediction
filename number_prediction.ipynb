# Predicting handwritten data using cnn method

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
from keras.layers import Dropout,Dense, Flatten,Conv2D,MaxPooling2D
from keras.models import Sequential
import pandas as pd
```

```python
#reading mnist data and splitting into test and train
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.mnist.load_data()
```

```python
#normalizing data
x_train=tf.keras.utils.normalize(x_train, axis=1)
x_test=tf.keras.utils.normalize(x_test, axis=1)
```

```python
#reshaping data for modelling
xtrain=x_train.reshape(-1,28,28,1)
xtest=x_test.reshape(-1,28,28,1)
```

```python
# building the model
model=Sequential()
model.add(Conv2D(64, (3,3), input_shape = xtrain.shape[1:], activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.compile(loss="sparse_categorical_crossentropy", optimizer='adam',metrics=['accuracy'])
model.fit(xtrain,y_train,epochs=10,validation_split=0.3)
```

```python
#predicting values for test data
predictions=model.predict(xtest)
predict=[np.argmax(x) for x in predictions]
df=pd.DataFrame({"actual_test":y_test,"predicted_test":predict})
```
